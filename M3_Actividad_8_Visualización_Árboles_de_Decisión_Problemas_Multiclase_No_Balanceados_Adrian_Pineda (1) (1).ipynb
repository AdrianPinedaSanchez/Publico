```python
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
# Prepare the data data
iris = datasets.load_breast_cancer()
X = iris.data
y = iris.target
# Fit the classifier with default hyper-parameters
clf = DecisionTreeClassifier(random_state=1234)
model = clf.fit(X, y)
```

```python
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
iris.target_names
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

```python
fig.savefig("decistion_tree_cancer.png")
```

## **3.- Investiga qué es un problema multiclase no balanceado. Descríbelo y muestra un ejemplo.**

Un problema multiclase no balanceado basicamente consiste a que las condiciones del problema se basan en que las clases de un dataset tienen un problema en torno a proporcion en su represetacion (algunas tienen una cantidad mucho mayor de datos comparadas con otras.

La complejidad de este problema para el aprendizaje automatico radica principalmente en la deteccion de las clases minoritarias en torno a su ajuste y balanco con las mayoritarias.

**Por ejemplo:** 

El diagnóstico de enfermedades raras. Supongamos que un conjunto de datos de diagnóstico médico tiene tres clases: "enfermedad A", "enfermedad B" y "enfermedad C". Si la "enfermedad A" es rara y solo está presente en el 1% de las muestras, mientras que las enfermedades "B" y "C" son más comunes y representan el 49% y el 50% de las muestras, respectivamente, el conjunto de datos se considera desbalanceado. 

Si un modelo de clasificación se entrena en este conjunto de datos, es probable que tenga dificultades para detectar la enfermedad A debido a su bajo número de muestras, lo que podría llevar a resultados inexactos o sesgados hacia las enfermedades más comunes.
